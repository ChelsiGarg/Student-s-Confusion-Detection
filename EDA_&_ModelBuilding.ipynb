```python
import pandas as pd

# Split data in training and testing set

train_n_conf=n_conf.iloc[0:140]
test_n_conf=n_conf.iloc[140:176]
train_conf=conf.iloc[0:147]
test_conf=conf.iloc[147:184]
train=[train_n_conf,train_conf]
train=pd.concat(train,sort=False)
test=[test_n_conf,test_conf]
test=pd.concat(test,sort=False)
train.shape
test.shape
```

```python
#Remove last column from data
ytrain=train.iloc[:,112]
ytrain.shape

ytest=test.iloc[:,112]
ytest.shape

train=train.iloc[:,:-1]
test=test.iloc[:,:-1]
```

```python
import numpy as np
import keras
from tensorflow.keras.utils import to_categorical

#one hot encoding
ytrain=keras.utils.np_utils.to_categorical(ytrain)
ytest=keras.utils.np_utils.to_categorical(ytest)

#converting to numpy array
train=np.asarray(train)
train=train.reshape(287,112,1) 

test=np.asarray(test)
test=test.reshape(73,112,1)
```

```python
import tensorflow as tf
import keras

# Build model by adding layers
def build_model():
  model=tf.keras.models.Sequential()
  #Layer 1
  model.add(tf.keras.layers.Conv1D(filters=64, kernel_size =8,activation='relu',padding='valid',input_shape=(112,1)))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.MaxPooling1D(padding='valid'))
  #Layer 2
  model.add(tf.keras.layers.Conv1D(filters=32,kernel_size =8,activation='relu',padding='valid',input_shape=(105,1) ))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.MaxPooling1D(padding='valid'))
  #Layer 3
  model.add(tf.keras.layers.Conv1D(filters=16,kernel_size =8,activation='relu',padding='valid',input_shape=(98,1) ))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.MaxPooling1D(padding='valid'))

  #Flatten Layer
  model.add(tf.keras.layers.Flatten())

  #DenseLayer1
  model.add(tf.keras.layers.Dense(25,activation='relu'))

  #Output Layer
  model.add(tf.keras.layers.Dense(2,activation='softmax' ))
  model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
  return model

model=build_model()
model.summary()
```

```python
# Train model
history=model.fit(train,ytrain,validation_split=0.25,epochs=6,batch_size=16)
```

```python
# Plot training and validation graphs
pd.DataFrame(history.history)
pd.DataFrame(history.history)[['accuracy','val_accuracy']].plot()
pd.DataFrame(history.history)[['loss','val_loss']].plot()
```

```python
# Test model with testing set
loss,accuracy=model.evaluate(test,ytest)
print('loss=',loss)
print('accuracy=',accuracy)
```

```python
P=model.predict(test)
print('P=',P)
print('shape of P=',P.shape)
```

```python
# Convert predictions to labels/classes
ypred=np.argmax(P,axis=1)
print('shape of ypred=',ypred.shape)
```

```python
ytest=np.argmax(ytest,axis=1)
print('shape of ytest=',ytest.shape)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Plot confusion matrix

cm=confusion_matrix(ytest,ypred)
np.set_printoptions(precision=1)
print(cm)
```

```python
cm_df=pd.DataFrame(cm,columns=['predicted negative','predicted positive'],index=['Actual negative','Actual positive'])

cm_df
```

```python
def binary_class_parameters(conf_matrix):
  TP=conf_matrix[1][1]
  TN=conf_matrix[0][0]
  FP=conf_matrix[0][1]
  FN=conf_matrix[1][0]

  print('True positives:',TP)
  print('True negatives:',TN)
  print('False positives:',FP)
  print('False negatives:',FN)

  ACC = (float(TP+TN)/float(TP+TN+FP+FN))
  MC=1-ACC
  RC=(TP/float(TP+FN))
  SP=(TN/float(TN+FP))
  PR=(TP/float(TP+FP))
  F1=2*((PR*RC)/(PR+RC))

  print('1. Accuracy : ',ACC)
  print('2. Mis-Classification : ',MC)
  print('3. Sensitivity recall : ',RC)
  print('4. Specificity : ',SP)
  print('5. Precision : ',PR)
  print('6. f_1 Score : ',F1)
  
```

```python
binary_class_parameters(cm)
```
